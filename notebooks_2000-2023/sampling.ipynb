```python
import pandas as pd 
YEAR_MIN = 2000
YEAR_MAX = 2023
N_PER_YEAR = 100 # can change to 50 if high compute

meta_path = "/Users/mariaworkman/fashion/fashion-neutrality/data/raw/VogueRunway.parquet"
df = pd.read_parquet(meta_path)

df2 = df[(df["year"] >= YEAR_MIN) & (df["year"] <= YEAR_MAX)].copy()

# Keep only actual runway looks (avoids backstage/detail shots)
df2 = df2[df2["section"].fillna("") == "Collection"]

# Require image URL
df2 = df2[df2["url"].notna()].copy()

sampled = (
    df2[df2["year"].between(2000, 2023)]
      .groupby("year", group_keys=False)
      .sample(n=N_PER_YEAR, random_state=0, replace=False)
      .reset_index(drop=True)
)

sampled.shape
counts = sampled["year"].value_counts().sort_index()
counts

```

now sample the collections

```python
import os

os.makedirs("../data/samples", exist_ok=True)
keep_cols = ["key", "designer", "season", "year", "category", "city", "section", "tags", "aesthetic", "url", "filename"]
keep_cols = [c for c in keep_cols if c in sampled.columns]

out_path = f"../data/samples/sample_{YEAR_MIN}_{YEAR_MAX}_{N_PER_YEAR}peryear.csv"
sampled[keep_cols].to_csv(out_path, index=False)
out_path
```
